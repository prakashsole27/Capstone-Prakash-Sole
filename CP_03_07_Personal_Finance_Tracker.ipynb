# Personal Finance Tracker using Python

This notebook helps track income, expenses, savings, and visualize spending patterns.

```
import pandas as pd
import matplotlib.pyplot as plt
```

```
data = {
    'Category': ['Salary', 'Food', 'Rent', 'Shopping', 'Transport', 'Entertainment'],
    'Amount': [50000, -5000, -15000, -3000, -2000, -2500]
}

df = pd.DataFrame(data)

print(df)
```

```
income = df[df['Amount'] > 0]['Amount'].sum()
expenses = abs(df[df['Amount'] < 0]['Amount'].sum())
savings = income - expenses

print(f"Total Income: ₹{income}")
print(f"Total Expenses: ₹{expenses}")
print(f"Total Savings: ₹{savings}")
```

```
expense_df = df[df['Amount'] < 0]

plt.figure(figsize=(8,5))
plt.pie(abs(expense_df['Amount']), labels=expense_df['Category'], autopct='%1.1f%%')
plt.title("Expense Distribution")
plt.show()
```
