# AI Meeting Summarizer using NLP

This notebook demonstrates an AI-powered meeting summarizer using Natural Language Processing.

```
# Install required libraries
# !pip install transformers torch
```

```
from transformers import pipeline
```

```
summarizer = pipeline("summarization", model="facebook/bart-large-cnn")
```

```
meeting_text = '''
Today we discussed the quarterly sales report, project deadlines,
marketing strategies, and upcoming client meetings.
The development team will complete the AI module by next week.
The marketing team will prepare social media campaigns.
'''
```

```
summary = summarizer(meeting_text, max_length=60, min_length=20, do_sample=False)

print("Meeting Summary:\n")
print(summary[0]['summary_text'])
```
